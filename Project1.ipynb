```python
########## QUESTION 2 ##########

def bisection(f, a, b, tol=1e-3, max_iter=100):
    fa = f(a)
    fb = f(b)
    if fa * fb > 0:
        raise ValueError("f(a)*f(b) should be <0 .")
    fx=1
    while abs(fx) > tol:
        c = (a + b) / 2
        fc = f(c)
        if fa * fc < 0:
            b = c
            fb = fc
        else:
            a = c
            fa = fc
        fx=fc
    return (c)

def newton_method(f, df, x0, tol=1e-10):
    
    x = x0
    xp = x0
    fx = f(x)
    i = 0
    print("\nn\txn\t\tf(xn)\t\terror\t\terror ratio")
    print("-\t-----\t\t-------\t\t-------\t\t----------")
    while abs(fx) > tol:
        x = x - fx / df(x)
        fx = f(x)
        i += 1
        error = abs(x-xp)
        err=error/abs(xp)
        print(i,"\t{:.10f}\t{:.10f}\t{:.10f}\t{:.10f}\n".format(x,fx,error,err))
        xp=x
        
    if abs(fx) <= tol:
        return x
    else:
        print("Failed to converge")
        
        
def f(x):
    return x**3 - 6

def df(x):
    return 3*x*x
    
a = 0
b = 20

xb = bisection(f, a, b,tol=1e-3)
print("\n Bisection Method Root = ",xb)

root = newton_method(f, df, xb)
print("\n Newton's Method Root = ",root)
```
